# Listas de Python 3
1. Es una estructura de datos Líneal de 0 o n elementos.
1. "Equivalente" a los arreglos.
1. Pueden ser de diferente tipo de dato [1,2,32,12] , [1, true, 2.34, "Hola"].
1. Pueden ser anidados, listas de listas. ["Melón", ["manzana roja", "manzana amarilla"].
1. Su selector es de la forma: [int].
1. Son Muteables.

```
# Ejemplos
numeros = [4,2,45,7,83,3]
print(numeros)
print(numeros[2])

# Una lista de 0 elementos
colores=[] # list()
type(colores) #averiguar de que tipo de dato es una variable con la funcion type
días=[]

print(colores)
print(días)
```

```
# Pueden de diferentes tipo de datos
mi_lista= [1,2,3,"José",False, 1.76,[552343455,453125452,556467465]]
print(mi_lista)
print(mi_lista[5])
print(mi_lista[6][1])
print(mi_lista[3][3])
print(mi_lista[6][1])
mi_lista= [1,2,3,"José",False, 1.76,["552343455","453125452","556467465"]]
print(str(mi_lista[6][1][2]))
```

```
frutas=["melon","sandia","kiwi","aguacate","mango","platano"]
print(frutas)
frutas[0]="Melón"
print(frutas)
#frutas[1][0]="S"
frutas[1]=frutas[1].capitalize()
print(frutas)
```

```
frutas=["melon","sandia","kiwi","aguacate","mango","platano"]
print(frutas)
frutas[0]="Melón"
print(frutas)
#frutas[1][0]="S"
frutas[1]=frutas[1].capitalize()
print(frutas)
####
####
# Las listas son clases por lo tanto, tienen métodos
#ejemplo: Append, agregar un elemento al final
frutas.append("Fresa")
print(frutas)
print("-----------")

print("Ejemplos de copias")
otras_frutas=frutas.copy()
print(frutas)
print(otras_frutas)
frutas[0]="Chicozapote"
print(frutas)
print(otras_frutas)
print("-----------")

print("ejemplo de extend")
colores=["rojo","verde","azul"]
frutas.extend(colores)
print(frutas)
print("-----------")


print("Ejemplo insert")
frutas.insert(3,"Limón")
print(frutas)
print("-----------")


print("Ejemplo remove, eliminar el kiwi")
frutas.remove("kiwi")
print(frutas)
print("-----------")


print("Ejemplo con operador del")
del frutas[2] # operador  de listas
print(frutas)

print(" ")


print("Ejemplo de clear")
frutas.clear()
print(frutas)
```
